```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load data
df = pd.read_csv('D:/lokasi_gerai_kopi_clean.csv')


# Pilih fitur numerik
features = ['x', 'y', 'population_density', 'traffic_flow', 'competitor_count', 'is_commercial']
X = df[features]

# Normalisasi
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(X_scaled)

# Visualisasi hasil cluster (berdasarkan lokasi x, y)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='kmeans_cluster', palette='tab10', s=60)
plt.title("Hasil Clustering dengan KMeans (n_clusters=5)")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Terapkan Agglomerative Clustering (default linkage='ward')
agglo = AgglomerativeClustering(n_clusters=5)
df['agglo_cluster'] = agglo.fit_predict(X_scaled)

# Visualisasi hasil cluster berdasarkan koordinat lokasi
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='agglo_cluster', palette='Set2', s=60)
plt.title("Hasil Clustering dengan Agglomerative Clustering")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Terapkan DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)  # kamu bisa eksperimen nilai eps dan min_samples
df['dbscan_cluster'] = dbscan.fit_predict(X_scaled)

# Visualisasi hasil DBSCAN
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='dbscan_cluster', palette='tab20', s=60)
plt.title("Hasil Clustering dengan DBSCAN")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.legend(title="Cluster (DBSCAN)")
plt.grid(True)
plt.show()
```
